## Named Entity Recognition (NER)

Named Entity Recognition (NER) is a Natural Language Processing (NLP) task used to identify and classify important entities in text.

These entities typically include names of **people, organizations, locations, dates, money, and other proper nouns**.

### Example

Sentence:

Elon Musk founded SpaceX in 2002 in California.

| Entity | Label |
|------|------|
| Elon Musk | PERSON |
| SpaceX | ORGANIZATION |
| 2002 | DATE |
| California | LOCATION |

```python
sentences = "In March 2024, Satya Nadella, the CEO of Microsoft, announced a new collaboration with OpenAI during a technology conference held in San Francisco, California. The event was attended by several prominent figures including Sundar Pichai from Google and Tim Cook from Apple. During the conference, Nadella explained that Microsoft plans to invest more than $10 billion into artificial intelligence research over the next five years. Researchers from Stanford University and the Massachusetts Institute of Technology also presented their latest work on machine learning models capable of understanding complex human language. Later that evening, the participants gathered at the Hilton Hotel near Union Square to discuss the future of technology and the role that AI will play in global industries by 2030."
```

```python
import nltk 
words =nltk.word_tokenize(sentences)
```

```python
nltk.download("maxent_ne_chunker")
nltk.download("words")
```

```python
tag_elements = nltk.pos_tag(words)
```

```python
nltk.ne_chunk(tag_elements).draw()
```
