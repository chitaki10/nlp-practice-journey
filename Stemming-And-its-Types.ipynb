```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()
test_words = [
    # Plurals
    "cats", "dogs", "horses", "buses", "wishes",
    
    # Verb tenses  
    "running", "ran", "running", "jumped", "swimming",
    
    # Derivations
    "program", "programs", "programmer", "programming", "programmed",
    
    # Edge cases (Porter rules)
    "caresses", "flies", "dies", "mules", "denied",
    "agreement", "fairly", "fairly", "sporting", "better",
    
    # NLP classics
    "studies", "happy", "happiness", "caring", "university"
]
```

### PorterStemmer


```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
for word in test_words:
    stemmed = stemming.stem(word)
    print(f"{word:12} → {stemmed}")
```

### Disadvantages in Stemming
- cant work on standatrad or complicated words 




```python
print(stemming.stem("congratulations"))
print(stemming.stem("tuticles"))
```

### RegexpStemmer

```python
from nltk.stem import RegexpStemmer

st = RegexpStemmer(r'ing$|s$|ed$|able$', min=4)

test_words = ["running", "cats", "jumped", "workable", "mass", "bee", "studies"]
for word in test_words:
    print(f"{word:10} → {st.stem(word)}")

```

### SnowballStemmer

```python
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer("english")

for word in test_words:
    p =snowball.stem(word)
    print(p)
```
