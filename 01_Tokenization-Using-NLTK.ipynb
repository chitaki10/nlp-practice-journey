```python
corpus = """The sent_tokenize function from the Natural Language Toolkit (NLTK) identifies sentence boundaries by using predefined punctuation rules and patterns. Here’s a breakdown of how it works:

Punctuation Marks: sent_tokenize looks for common sentence-ending punctuation such as periods (.), exclamation marks (!), and question marks (?). It recognizes these marks as indicators that one sentence has ended and another may begin.

Whitespace and Newlines: It also considers whitespace characters (like spaces and newlines) following punctuation marks to determine where sentences end.

Language Support: The tokenizer is designed to handle different languages and takes into account specific linguistic patterns that might be relevant in those languages.

Contextual Information: Understanding sentence boundaries is crucial for certain Natural Language Processing (NLP) tasks such as named entity recognition or dependency parsing, where knowing the beginning and end of sentences is important for context.

Usage in Frameworks: The function can be conveniently used in various applications where text needs to be broken down into more manageable pieces, allowing further analysis, like tokenizing into words afterward with a word_tokenize function.

In practice, when you apply sent_tokenize to a string of text, it returns a list of sentences by splitting the text at these identified boundaries, helping to structure the data effectively for further processing tasks. If you have any further questions about its application or need examples, feel free to ask!

Was this content relevant to you?
 """
```

```python
print(corpus)
```

### Sentence ---> paragraphs

```python
##Tokenization 
##Sentence ---> paragraphs

from nltk.tokenize import sent_tokenize
document = sent_tokenize(corpus)
for sentence in document:
    print(sentence)
```

### Paragraph --->words

```python
from nltk.tokenize import word_tokenize
for sentence in document:
    print(word_tokenize(sentence))
```

### wordpunct_tokenize
-  Word + punctuation as separate tokens—aggressive split.
Remember: "WordPUNCT"—everything gets chopped. Good for regex-heavy preprocessing.
Production fail: Over-splits OOV words → embedding lookup fails.

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

### TreebankWordTokenizer
- Splits on punctuation + spaces, treats contractions as single tokens (e.g., "don't" → ['don', "'t"]).
- Remember: "Bank" rules—smart punctuation handling for Penn Treebank corpus. Production fail: Misses apostrophes → broken sentiment models.

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
