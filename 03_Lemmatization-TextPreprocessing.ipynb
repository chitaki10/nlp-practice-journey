## Lemmatization

Lemmatization is a text preprocessing technique that converts words to their **dictionary base form (lemma)** using linguistic knowledge.

Example:

| Word | Lemma |
|-----|------|
| running | run |
| better | good |
| studies | study |

Unlike stemming, lemmatization produces **meaningful words**.

```python
test_words = [
    # Plurals
    "cats", "dogs", "horses", "buses", "wishes",

    # Verb tenses
    "running", "ran", "running", "jumped", "swimming",

    # Derivations
    "program", "programs", "programmer", "programming", "programmed",

    # Edge cases (Porter rules)
    "caresses", "flies", "dies", "mules", "denied",
    "agreement", "fairly", "fairly", "sporting", "better",

    # NLP classics
    "studies", "happy", "happiness", "caring", "university"
]
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("going",pos='v')
```

```python
for word in test_words:
  print( lemmatizer.lemmatize(word, pos='v'))
```
