```python
paragraph = "Artificial Intelligence has rapidly transformed the way humans interact with technology over the past decade. Companies such as Google, Microsoft, and OpenAI are investing billions of dollars into research and development to create systems that can understand language, recognize images, and make intelligent decisions. In 2023, many organizations began integrating large language models into their applications, enabling computers to answer questions, summarize documents, and even write code. Although these systems are powerful, researchers continue to debate the ethical implications of automation, data privacy, and algorithmic bias. As technology evolves, society must carefully balance innovation with responsibility to ensure that artificial intelligence benefits everyone."
```

```python
print(paragraph)
```

```python
# we will find the pos tag 
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
# we will find the pos tag 
import nltk
for i in range(len(paragraph)):
    words = nltk.word_tokenize(paragraph[i])
    words = [word for word in words]
    poss_tag = nltk.pos_tag(words)
    print(poss_tag)
```
